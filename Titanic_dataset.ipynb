Data Visualization

```python
sns.distplot(data['Survived'])
```

```python
sns.scatterplot(x='Age',y='PassengerId',data=data)
plt.show()
```

```python
sns.countplot(x='PassengerId',data=data, hue='Age')
plt.show()
```

```python
sns.countplot(x='Survived',data=data, hue='Pclass')
plt.show()
```

```python
sns.countplot(x='Ticket',data=data, hue='Pclass')
plt.show()
```

```python
sns.violinplot(y='PassengerId',x='Survived',data=data, hue='Age')
plt.show()
```

```python
sns.pairplot(data=data, hue='Age')
plt.show()
```

```python
sns.pairplot(data=data, hue='Age',palette='mako')
plt.show()
```

```python
sns.boxplot(x='SibSp',y='Age',data=data,palette='mako')
plt.show()
```
